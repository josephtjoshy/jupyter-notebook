```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
from sklearn import linear_model
```

```python
data=pd.read_csv("data.csv")
data
```

```python
plt.scatter(x=data['X'],y=data['Y'])
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[['X']],data[['Y']])
```

```python
reg.predict([[65]])
```

```python
bias=reg.intercept_
weight=reg.coef_
m=[10,20,30,45,50,60,70,25,26,35,100]
y=[]
for i in m:
    y.append((i*weight)+bias)
```

```python
yy=[]
for i in y:
    for j in i:
        for k in j:
            yy.append(k)
yy
```

```python
plt.scatter(x=data['X'],y=data['Y'],color='red')
plt.plot(data['X'],yy,color='blue')
```
